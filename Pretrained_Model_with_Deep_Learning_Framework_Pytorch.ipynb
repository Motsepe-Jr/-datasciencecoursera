```
from torchvision import models
```

```
# let take a look at some pretained models
dir(models)
```

```
# Lets run an instance of AlexNet class
alexnet = models.AlexNet()
```

```
# Lets instantiate Resnet Weights
resnet = models.resnet101(pretrained=True) # 44.5 million parameters
resnet
```

```
# Preprocessing functios for an image
from torchvision import transforms
preprocess = transforms.Compose([
                                 transforms.Resize(256),
                                 transforms.CenterCrop(224),
                                 transforms.ToTensor(),
                                 transforms.Normalize(
                                     mean=[0.485, 0.456, 0.406],
                                     std=[0.229, 0.224, 0.225]
                                 )
])
```

```
from PIL import Image
img = Image.open('dog.jpg')
```

```
img_t = preprocess(img)
```

```
img_t.shape
```

```
import torch
batch_t = torch.unsqueeze(img_t, 0)

batch_t.shape
```

```
resnet.eval()
```

```
out = resnet(batch_t)
out
```

```
with open('labels.txt') as f:
  labels = [line.strip() for line in f.readlines()]
```

```
# let get the maximum index from output
_, index = torch.max(out, 1)
index
```

```
percentage = torch.nn.functional.softmax(out, dim=1)[0] * 100
labels[index[0]], percentage[index[0].item()]
```
